```python
import re 
```

```python
from uuid import uuid4
from datetime import datetime 

from app_helper import (
        escape_single_quote, 
        df_to_csv, 
        list2sql_str, 
        DBConn, )
```

```python
filename = "dev/cornell-awards.txt"
```

```python
with open(filename, encoding="utf-8") as f:
    data = [i.strip() for i in f.read().split("\n") if i.strip()]
```

```python
data
```

```python
x = '2014: MIT Energy Initiative Award'
print(re.match("\d{4}:", x))
```

```python
person_award_dict = {}
year_award_list = []
awards = {}
person_name = data[0]
for x in data[1:]:
    if re.match("\d{4}:", x):
        y = x.split(": ")
        year_award_list.append(y)
        awards[y[1].strip()] = True 
    else:
        if year_award_list:
            person_award_dict[person_name] = year_award_list
            year_award_list = []
        person_name = x
# add last one
if year_award_list:
    person_award_dict[person_name] = year_award_list
```

```python
award_list = sorted(awards.keys())
```

```python
len(award_list), award_list
```

```python
_ts = str(datetime.now())
with DBConn() as _conn:
    for x in award_list:
        sql = f"""
        insert into g_entity(id,uid,ts,entity_type,name)
        values ('{str(uuid4())}', '0', '{_ts}','award','{escape_single_quote(x)}');
        """
        print(sql)
        _conn.execute(sql)
        
    _conn.commit()
```

```python
award_data = []
for k,v in person_award_dict.items():
    for i in v: 
        prop = f"org:Cornell, year:{i[0]}"
        award_data.append([k, i[1], prop])
```

```python
len(award_data), award_data
```

```python
_ts = str(datetime.now())
with DBConn() as _conn:
    for x in award_data:
        person_name, award_name, props=x
        sql = f"""
        insert into g_relation(id,uid,ts,rel_type
        ,ref_tab,ref_key,ref_val
        ,ref_tab_sub,ref_key_sub,ref_val_sub
        ,props)
        values ('{str(uuid4())}', '0', '{_ts}','person-award'
        ,'g_person','name','{escape_single_quote(person_name)}'
        ,'g_entity','name','{escape_single_quote(award_name)}'
        ,'{props}');
        """
        print(sql)
        _conn.execute(sql)
        
    _conn.commit()
```
