```python
class DBConn(object):
    def __init__(self):
        if FILE_DB.endswith("duckdb"):
            self.conn = duckdb.connect(FILE_DB)
        else:
            self.conn = sqlite3.connect(FILE_DB)

    def __enter__(self):
        return self.conn

    def __exit__(self, type, value, traceback):
        self.conn.close()
```

```python
FILE_DB = "../db/cs-faculty-20230429.duckdb"
```

```python
with DBConn() as _conn:
    alter_sql = """
        ALTER TABLE g_task drop due_date;
        ALTER TABLE g_task add column due_date DATE;
    """
    df = _conn.execute(alter_sql).df()
    df
```

```python
with DBConn() as _conn:
    alter_sql = """
        ALTER TABLE g_task drop alert_date;
        ALTER TABLE g_task add column alert_time TIMESTAMP;
    """
    df = _conn.execute(alter_sql).df()
    df
```

```python
with DBConn() as _conn:
    sql_stmt = """
    select  c.table_name, c.column_name, c.data_type --, c.*
    from information_schema.columns c
    where 1=1
    -- and c.table_catalog like 'faculty-Cornell-CS%'
    and c.table_schema = 'main'
    and c.table_name like 'g_%' and c.table_name = 'g_task'
    order by c.table_name, c.ordinal_position
    ;
    """    
    df = pd.read_sql(sql_stmt, _conn)
```
