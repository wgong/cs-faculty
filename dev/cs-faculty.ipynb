```python
def _load_db(DEV_FLAG=True):
    FILE_ROOT = "../app/db" if DEV_FLAG else "./db"
    FILE_DB = f"{FILE_ROOT}/faculty-Cornell-CS.duckdb"
    FILE_XLSX = f"{FILE_ROOT}/faculty-Cornell-CS.xlsx"

    if not Path(FILE_DB).exists():
        if not Path(FILE_XLSX).exists():
            raise Exception(f"source file: {FILE_XLSX} missing")
        
        xls = pd.ExcelFile(FILE_XLSX)
        sheet_name = "Faculty"
        df_faculty = pd.read_excel(xls, sheet_name, keep_default_na=False)
        sheet_name = "Research Groups"
        df_research_group = pd.read_excel(xls, sheet_name, keep_default_na=False)

        conn = dd.connect(FILE_DB)
        conn.register("v_faculty", df_faculty)
        conn.register("v_research_group", df_research_group)

        conn.execute("Create table t_faculty as select * from v_faculty;")
        conn.execute("Create table t_research_group as select * from v_research_group;")
        conn.close()

    else:
        conn = dd.connect(FILE_DB)
        df_faculty = conn.execute("select * from t_faculty;").fetchdf()
        df_research_group = conn.execute("select * from t_research_group;").fetchdf()

    return df_faculty, df_research_group
```

```python
df_faculty_, df_research_group_ = _load_db()
```
